```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt

inputExcelFile ="User study.xlsx"

df = pd.read_excel(inputExcelFile, sheet_name="Sheet1")



df.rename(columns={
    df.columns[1]: "starttime",
    df.columns[2]: "endtime",
    df.columns[5]: "version",
    df.columns[6]: "userid",
    df.columns[7]: "isExpert",
    df.columns[8]: "Q1",
    df.columns[9]: "Q2",
    df.columns[10]: "Q3",
    df.columns[11]: "Q4",
    df.columns[12]: "Q5",
    df.columns[13]: "Q6",
    df.columns[14]: "Q7",
    df.columns[15]: "Q8",
    df.columns[16]: "Q9",
    df.columns[17]: "Q10",
    df.columns[18]: "Q11",
    df.columns[19]: "Q12",
    df.columns[20]: "Q13",
    df.columns[21]: "Q14",
    df.columns[22]: "Q15",
    df.columns[23]: "Q16",
    df.columns[24]: "Q17",
    df.columns[25]: "Q18",
    df.columns[26]: "Q19",
    df.columns[27]: "Q20",
}, inplace=True)

df.drop(columns=[df.columns[1],df.columns[2],df.columns[3], df.columns[4]], inplace=True)

df["isExpert"] = df["isExpert"].apply(lambda x: "Non-Expert" if x == "No" else "Expert")

#userID column only get userID using regex. sometimes could contain 'User ID: XXXX...' where XX.. is the actual userID
df["userid"] = df["userid"].apply(lambda x: x.split(":")[1].strip() if "User ID:" in x else x)

#rename versions: 1 --> LC + EH, 2 --> LC + ES, 3 --> HC + EH, 4 --> HC + ES
df["version"] = df["version"].apply(lambda x: "LC + EH" if x == 1 else "LC + ES" if x == 2 else "HC + EH" if x == 3 else "HC + ES")


df.to_csv("questionnaire_parsed.csv", index=False)


df
```

```python
# create a boxplot for each question (Q1 - Q15) per version.


melted_data = df.melt(id_vars=['version', 'userid', 'isExpert'], value_vars=df.columns[3:], var_name='question')

#remove Q16 - Q20
melted_data = melted_data[melted_data["question"].apply(lambda x: "Q" in x and int(x[1:]) <= 15)]

```

```python
chart = alt.Chart(melted_data).mark_rect().encode(
    x=alt.X("version", sort=["LC + EH", "LC + ES", "HC + EH", "HC + ES"]),
    y=alt.Y("question:N", sort=["Q1", "Q2", "Q3", "Q4", "Q5", "Q6", "Q7", "Q8", "Q9", "Q10", "Q11", "Q12", "Q13", "Q14", "Q15"]),
    color=alt.Color('value:Q', scale=alt.Scale(scheme='redblue', domain=[0, 10])),
).properties(
    width=600,
    height=400
)
chart
```

```python
likert= df[df["version"] == "HC + ES"].melt(id_vars=['version', 'userid', 'isExpert'], value_vars=df.columns[3:], var_name='question')
#remove Q16 - Q20 and Q8
likert = likert[likert["question"].apply(lambda x: "Q" in x and int(x[1:]) <= 15)]
likert = likert[likert["question"] != "Q8"]

likert = likert.replace({
    "value": {
        0: 0,
        1: 0,
        2: 1,
        3: 1,
        4: 2,
        5: 3,
        6: 4,
        7: 5,
        8: 5,
        9: 6,
        10:6
    }
})

likert = likert.groupby(["question", "value"]).count().reset_index()[["question", "value", "userid"]].rename(columns={"userid": "count"})

likert = likert.sort_values(by=['value'], ascending=False)

#make a red-blue color scale

color_scale = alt.Scale(
    scheme='redblue',
)

hc_es = alt.Chart(likert).mark_bar().encode(
    x=alt.X('sum(count)', stack="normalize"),
    y=alt.Y("question:N", sort=["Q1", "Q2", "Q3", "Q4", "Q5", "Q6", "Q7", "Q8", "Q9", "Q10", "Q11", "Q12", "Q13", "Q14", "Q15"]),
    color=alt.Color('value:N',scale= color_scale, legend=alt.Legend(title="Likert scale"))
)

# hc_es
likert
```

```python
#count number of experts and non-experts for each version

counted = df.groupby(["version", "isExpert"]).count().reset_index()[["version", "isExpert", "userid"]].rename(columns={"userid": "count"})




x = alt.Chart(counted).mark_bar().encode(
    x=alt.X("version", sort=["LC + EH", "LC + ES", "HC + EH", "HC + ES"]),
    y=alt.Y('count:Q', scale=alt.Scale(nice=1)),
    xOffset="isExpert:N",
    color=alt.Color('isExpert:N', title='User type'),
).properties(
    title="Number of experts and non-experts for each version"
)
x
```

```python
#Shows preference for barchart vs feedback image

q8 = df[["isExpert", "Q8"]].groupby("isExpert").mean().reset_index().rename(columns={"Q8": "score"})


alt.Chart(q8).mark_bar().encode(
    x=alt.X('score:Q',
            scale=alt.Scale(domain=[-10, 10]),
            axis=alt.Axis(values=[-10, 10],
                          labelExpr="datum.value === -10 ? 'Feedback image' : 'Barchart'")
           ),
    y=alt.Y('isExpert:N', sort=alt.EncodingSortField(field="isExpert", order="ascending")),
    color= alt.Color('isExpert:N', title='User type'),
).properties(
    width=400,
    height=100,
    title="Preference for barchart vs feedback image"
)
```

```python
#boxplot per construct.

hces = df[df["version"] == "HC + ES"].copy()
hces = df[['version', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q7', 'Q9', 'Q10', 'Q11','Q12','Q13','Q14', 'Q15']].copy()


constructs = {
    'Accuracy': ['Q1'],
    'Satisfaction': ['Q11'],
    'Explanation': ['Q2', 'Q7', 'Q9'],
    'Control': ['Q3','Q5'],
    'Usability': ['Q4', 'Q10'],
    'Trust': ['Q11', 'Q12'],
    'Use_Intention': ['Q14', 'Q15']
}

dflist = {
    'Accuracy': None,
    'Satisfaction': None,
    'Explanation': None,
    'Control': None,
    'Usability': None,
    'Trust': None,
    'Use_Intention': None
}

charts = []
for construct in constructs:
    cols = constructs[construct]
    cols.append("version")
    data = hces[cols].copy().melt(id_vars=['version'], value_vars=cols[:-1], var_name='question')
    plot = alt.Chart(data).mark_boxplot(ticks=True).encode(
    x=alt.Y('value:Q', scale=alt.Scale(domain=[0, 10]), title="Likert scale score"),
    y=alt.X('question:N', sort=["LC + EH", "LC + ES", "HC + EH", "HC + ES"]),
    color=alt.Color('version:N', title='Version'),
    row=alt.Row('version:N', align= 'each' ,title="Version", sort=["LC + EH", "LC + ES", "HC + EH", "HC + ES"], spacing=30)
    ).properties(
    title=construct,
    height=100
    )
    charts.append(plot)


charts[0] | charts[1] | charts[2] 
```
